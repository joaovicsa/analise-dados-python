```python
import pandas as pd
import numpy as np
```

```python
data = {
    'ID': range(1, 26),
    'Duration': [60]*25,
    'Date': ['2020/12/01', '2020/12/02', '2020/12/03', '2020/12/04', '2020/12/05', '2020/12/06', '2020/12/07', '2020/12/08', '2020/12/09', '2020/12/10', '2020/12/11', '2020/12/12', '2020/12/13', '2020/12/14', '2020/12/15', '2020/12/16', '2020/12/17', '2020/12/18', '2020/12/19', '2020/12/20', '2020/12/21', None, '20201226', '2020/12/24', '2020/12/25'],
    'Pulse': [110]*25,
    'Maxpulse': [130]*25,
    'Calories': [409.1, 479.0, 340.0, 282.4, 406.0, 300.0, 374.0, 253.3, 311.2, 606.2, 329.3, 250.7, 345.0, 375.6, 540.2, 451.1, 330.0, np.nan, 322.0, 318.0, 308.0, 320.5, 420.2, 311.8, 334.0]
}
df = pd.DataFrame(data)
```

```python
print("--- Informações Gerais ---")
df.info()
print("\n--- Primeiras 5 linhas ---")
print(df.head())
print("\n--- Últimas 5 linhas ---")
print(df.tail())
```

```python
df_clean = df.copy()
```

```python
df_clean['Calories'] = df_clean['Calories'].fillna(0)
print("\n--- Verificação: Calories (nulos substituídos por 0) ---")
print(df_clean.iloc[17:19])
```

```python
df_clean['Date'] = df_clean['Date'].fillna('1900/01/01')
print("\n--- Verificação: Date (nulos substituídos por 1900/01/01) ---")
print(df_clean.iloc[20:23])
```

```python
df_clean['Date'] = df_clean['Date'].replace('1900/01/01', np.nan)
```

```python
df_clean['Date'] = df_clean['Date'].replace('20201226', '2020/12/26')
```

```python
df_clean['Date'] = pd.to_datetime(df_clean['Date'])

print("\n--- Verificação: Date após conversão para Datetime ---")
print(df_clean.info())
```

```python
df_clean.dropna(subset=['Date'], inplace=True)
```

```python
print("\n--- Dataframe Final Limpo ---")
print(df_clean)
```
